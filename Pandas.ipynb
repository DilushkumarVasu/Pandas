```python
import pandas as pd
import matplotlib.pyplot as plt

# Example Data
data = pd.Series([35, 25, 20, 20], index=['Product A', 'Product B', 'Product C', 'Product D'])

# Pie Chart with autopct, explode, and pctdistance
data.plot.pie(
    autopct='%1.1f%%',          # Displays percentage on slices
    explode=None,   # Separates certain slices
    pctdistance=0.8,            # Adjusts the distance of the percentage from the center
    startangle=90,              # Start the pie chart at 90 degrees
    shadow=True,                # Adds shadow for 3D effect
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],  # Custom colors
    title='Product Sales Distribution'
)

plt.xlabel("")
plt.ylabel("")  # Removes the y-axis label for a cleaner look
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example Data
data = pd.Series([35, 25, 20, 20], index=['Product A', 'Product B', 'Product C', 'Product D'])

# Create a Donut Chart
data.plot.pie(
    autopct='%1.1f%%',                 # Displays percentage on slices
    explode=None,                      # No slices exploded
    pctdistance=0.8,                   # Adjusts distance of percentage from the center
    startangle=90,                     # Start the pie chart at 90 degrees
    shadow=True,                       # Adds shadow for 3D effect
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],  # Custom colors
    title='Product Sales Distribution',
    wedgeprops={'width': 0.3}          # Adjusts the width of slices to make a donut
)
plt.ylabel("")  # Optional: Removes the y-axis label for a cleaner look
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example Data
data = pd.Series([35, 25, 20, 20], index=['Product A', 'Product B', 'Product C', 'Product D'])

# Create Donut Chart
fig, ax = plt.subplots()
data.plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],
    title='Product Sales Distribution',
    wedgeprops={'width': 0.3, 'edgecolor': 'gray', 'linewidth': 1},
    ax=ax
)

# Add Text in the Center
center_text = "Total Sales\n100%"
ax.text(0, 0, center_text, ha='center', va='center', fontsize=12, weight='bold')

plt.ylabel("")  # Optional: Removes y-axis label for a cleaner look
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

T = 1000
ang = np.linspace(0, 2 * np.pi, T)

def circle(r,l):
  a = np.multiply(4 * np.random.random(T) + r, np.sin(ang))
  b = np.multiply(4 * np.random.random(T) + r, np.cos(ang))
  label = np.ones(T) * l
  return a,b,label

xi,yi,li = circle(5,0)
xj,yj,lj = circle(10,1)

'''
ang = np.linspace(0, 2 * np.pi, T)
xr5 = np.multiply(4 * np.random.random(T) + 5, np.sin(ang))
xr10 = np.multiply(4 * np.random.random(T) + 10, np.sin(ang))
l1 = np.zeros(T)

yr5 = np.multiply(4 * np.random.random(T) + 5, np.cos(ang))
yr10 = np.multiply(4 * np.random.random(T) + 10, np.cos(ang))
l2 = np.ones(T)
'''

X1 = np.concatenate((xi,xj),axis=0)
X2 = np.concatenate((yi,yj),axis=0)
Y = np.concatenate((li,lj),axis=0)
#plt.scatter(X1,X2,c=Y);

plt.scatter(X1, X2, c=Y)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter Plot of Circle Data")
plt.show()



data = {
    "X1": X1,
    "X2": X2,
    "Y": Y
}

df = pd.DataFrame(data)
#print(df)

#Create a dataframe with (x1,x2,y) first, then add x1^2,X2^2, and x1*x2

def squareX1(row):
    return row['X1']**2

def squareX2(row):
    return row['X2']**2

def product(row):
  return row['X1'] * row['X2']


#df['X1^2'] = df.apply(squareX1,axis=1)
df['X1^2'] = df.apply(lambda row:row['X1']**2,axis=1)

#df['X2^2'] = df.apply(squareX2,axis=1)
df['X2^2'] = df.apply(lambda row:row['X2']**2,axis=1)

#df['X1*X2'] = df.apply(product,axis=1)
df['X1*X2'] = df.apply(lambda row:row['X1']*row['X2'],axis=1)

df.to_csv('output.csv',header=False,index=False)
df.head()

```
